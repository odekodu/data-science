### What is statistics?
Statistics is the science of collecting, organizing and analyzing `data`, `for better decision making`.

What is data?
It is facts or pieces of information that can be `measured`.
    e.g: Measure the IQ of a class
    {98, 89, 99, 50, 94}


### Types of statistics
1. Descriptive statistics
    It consists of organizing and summerizing of data

2. inferential Statistics
    This is a technique where in we use the `data` that we have measured to form `conclusions`


Classroom of math students (20)
Marks = {90, 67, 78, 89, 67, 87, 66, 89, 88, 77, 87, ...}

e.g questions. `Descriptive stats`
What is the avg mark of the students?

e.g questions. `Inferential stats`
Are the marks of the students of this classroom similar to the marks of the maths classroom in the college?

Here college is the `population` and the classroom is the `sample`.


### Population(N) and Sample(n)
Population(N) is the entire data collection while Sample(n) is a selected piece of the data.
    e.g. Elections


### Sampling Techniques
This completely dependent of the use case. 
You can combine multiple techniques.

1. `Simple Random Sampling`: Every member of the population(N) has an equal chance of being selected for the sample(n).
    e.g. Election exit poll needs people that voted.
2. `Stratified Sampling`: This is the technique where the population (N) is split into `non-overlapping groups` called `strata`.
    e.g. Sampling based on gender, age group, profession
3. `Systematic Sampling`: Picks an nth item in a population(N).
4. `Convenience Sampling`: This technique picks only items that have interest in the topic. 
    e.g. Stackoverflow Surveys is for developers only.
    

### Variables
A variable is a property that can take on any value. e.g. Height, Age, Weight.

Types of variable
1. Quantitive variables
    They can be measured numerically 
    e.g Age, Height

Types of Quantitive variables
- Discrete Variables
    Basically whole numbers. e.g Bank Accounts, No. of students in a class.

- Continious Variables
    e.g Height, Weight

2. Qualitive/Categorical variables.
    Based on some characteritics we can derive qualitative variables
    e.g Gender, Profession, Age Groups, Blood Group, Cloth size.


### Variable Measurement Scales
There are four types of measured variables

1. Nominal data (Categorical Data)
    e.g Colors

2. Ordinal
    Here the order of the data matters but value doesn't.
    e.g Ranks

3. Interval
    Here value and order matters, natural zero is not present.
    e.g Temperature, Age Group.

4. Ratio
