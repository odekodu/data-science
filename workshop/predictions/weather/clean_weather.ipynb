### Data Cleaning

```python
# Remove the unwanted columns
df = df[['Time (hr)', 'Temperature (°c)', 'Forecast (°c)',
       'Rain (mm)', 'Rain (%)', 'Cloud (%)', 'Pressure (mb)', 'Wind (km/h)',
       'Gust (km/h)', 'Direction (deg)', 'Moonrise (hr)', 'Moonset (hr)',
       'Sunrise (hr)', 'Sunset (hr)', 'Weather', 'Date']]
```

```python
df.head()
```

```python
# Convert Time (hr) to int
df['Time (hr)'] = df['Time (hr)'].astype(int)
```

```python
# Convert Temperature (°c) to int
df['Temperature (°c)'] = df['Temperature (°c)'].str.replace(' °c', '').astype(int)
```

```python
# Convert Forecast (°c) to int
df['Forecast (°c)'] = df['Forecast (°c)'].str.replace(' °c', '').astype(int)
```

```python
# Convert Rain (mm) to float
df['Rain (mm)'] = df['Rain (mm)'].str.replace(' mm', '').astype(float)
```

```python
# Convert Rain (%) to float
df['Rain (%)'] = df['Rain (%)'].str.replace('%', '').astype(float)
```

```python
# Convert Cloud (%) to float
df['Cloud (%)'] = df['Cloud (%)'].str.replace('%', '').astype(float)
```

```python
# Convert Pressure (mb) to float
df['Pressure (mb)'] = df['Pressure (mb)'].str.replace(' mb', '').astype(float)
```

```python
# Convert Wind (km/h) to float
df['Wind (km/h)'] = df['Wind (km/h)'].str.replace(' km/h', '').astype(float)
```

```python
# Convert Gust (km/h) to float
df['Gust (km/h)'] = df['Gust (km/h)'].str.replace(' km/h', '').astype(float)
```

```python
df.head()
```

```python
df['Direction (deg)'] = df['Direction (deg)'].str.replace('transform: rotate(', '')
```

```python
df['Direction (deg)'] = df['Direction (deg)'].str.replace('deg);', '')
```

```python
df['Direction (deg)'] = df['Direction (deg)'].astype(float)
```

```python
df.head()
```

##### Convert [Moonrise, Moonset, Sunrise, Sunset] to minutes

```python
def hours_to_minutes(time_text: str):
    # Takes a time 'hh:mm'
    # Extracts the hours and minutes
    # Converts to minutes

    if time_text.count('No') > 0:
        return np.nan

    hours, minutes = map(int, time_text.replace('AM', '').replace('PM', '').split(":"))

    # Add 12 hours if PM
    hours = hours + 12 if time_text.count('PM') else hours
    minutes = (hours * 60) + minutes
    return minutes
```

```python
# Remove extra spaces from Moonrise (hr)
df['Moonrise (hr)'] = df['Moonrise (hr)'].str.strip()
```

```python
df['Moonrise (hr)'].unique()
```

```python
# Convert Moonrise (hr) to minutes
df['Moonrise (min)'] = [hours_to_minutes(x) for x in df['Moonrise (hr)']]
```

```python
# Convert Moonset (hr) to minutes
df['Moonset (min)'] = [hours_to_minutes(x) for x in df['Moonset (hr)']]
```

```python
# Convert Sunrise (hr) to minutes
df['Sunrise (min)'] = [hours_to_minutes(x) for x in df['Sunrise (hr)']]
```

```python
# Convert Sunset (hr) to minutes
df['Sunset (min)'] = [hours_to_minutes(x) for x in df['Sunset (hr)']]
```

##### Fill missing values

```python
df['Sunset (min)'].fillna(df['Sunset (min)'].mean())
```

```python
df['Sunrise (min)'].fillna(df['Sunrise (min)'].mean())
```

```python
df['Moonrise (min)'].fillna(df['Moonrise (min)'].mean())
```

```python
df['Moonset (min)'].fillna(df['Moonset (min)'].mean())
```

```python
df.head()
```

```python
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df.head()
```

```python
# Determine Season based on month
df['Season'] = ['wet' if x >= 4 and x <= 9 else 'dry' for x in df['Date'].dt.month]
```
