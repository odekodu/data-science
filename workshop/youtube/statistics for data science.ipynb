### What is statistics?
Statistics is the science of collecting, organizing and analyzing `data`, `for better decision making`.

What is data?
It is facts or pieces of information that can be `measured`.
    e.g: Measure the IQ of a class
    {98, 89, 99, 50, 94}


### Types of statistics
1. Descriptive statistics
    It consists of organizing and summerizing of data

2. inferential Statistics
    This is a technique where the `data` that have been measured is used to form `conclusions`.


`Imagine`: Classroom of math students (20)
Marks = {90, 67, 78, 89, 67, 87, 66, 89, 88, 77, 87, ...}

Descriptive stats question:
What is the avg mark of the students?

Inferential stats question:
Are the marks of the students of this classroom similar to the marks of the maths classroom in the college?

Here college is the `population` and the classroom is the `sample`.


### Population(N) and Sample(n)
Population(N) is the entire data collection while Sample(n) is a selected piece of the data.
    e.g. Elections


### Sampling Techniques
Selecting a sample technique is completely dependent of the use case. 
You can combine multiple techniques.

1. `Simple Random Sampling`: Every member of the population(N) has an equal chance of being selected for the sample(n).
    e.g. Election exit poll needs people that voted.
2. `Stratified Sampling (Group Sampling)`: This is the technique where the population (N) is split into `non-overlapping groups` called `strata`.
    e.g. Sampling based on gender, age group, profession
3. `Systematic Sampling`: Picks an nth item in a population(N).
4. `Convenience Sampling`: This technique picks only items that have interest in the topic. 
    e.g. Stackoverflow Surveys is for developers only.

### Measure of Dispersion. 
This is the measure of how `spread` a data set is.
It can be measure using variance and standard diviation


### Variance
This is the measure of the spread between numbers in a data set.
(Population (N) Variance)^2 = (sum(x - population mean)^2) / N
(Sample (n) Variance)^2 = (sum(x - sample mean)^2) / (n - 1)

`Note`:
Variance is higher when the data is more dispersed.


### Standard Diviation
It measures the variation of a random variable from the mean which is the center of the data.
std = sqrt(variance)

# Day 4: Outliers

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = [10, 11, 10, 13, 10, 11, 10, 13, 19, 19, 100, 11, 11, 13, 17, 11, 11, 13, 17, 15, 13, 10, 11, 15, 13, 14, 18, 104, 106]
```

```python
def detect_outliers(data):
    threshhold = 3
    mean = np.mean(data)
    std = np.std(data)
    outliers = []

    for i in data:
        z_score = (i - mean) / std
        print(z_score, i)
        if (np.abs(z_score) > threshhold):
            outliers.append(i)
    return outliers
```

```python
detect_outliers(data)
```
