```python
import pandas as pd
```

```python
df = pd.read_csv('../../../data/epl_matches.csv')
df.head()
```

```python
# Rename column name to team
df.rename(columns= { 'name': 'team' }, inplace= True)
```

```python
# Remove trailing spaces from the team name
df['team'] = df['team'].str.strip()
```

```python
# Convert date from string to pd.datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Create a simple predictor
df['venue_code'] = df['venue'].astype('category').cat.codes
```

```python
df.head()
```

```python
df['opp_code'] = df['opponent'].astype('category').cat.codes
df.head()
```

```python
df['hour'] = df['time'].replace(':.+', '', regex= True).astype('int')
df.head()
```

```python
df['day_code'] = df['date'].dt.day_of_week
df.head()
```

```python
df['target'] = (df['result'] == 'W').astype('int')
df.head()
```

```python
df['result'].astype('category').cat.codes
```

```python
# Creating the machine learning model
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators= 50, min_samples_split= 10)
```

```python
train = df[df['date'] < '2022-01-01']
```

```python
test = df[df['date'] >= '2022-01-01']
```

```python
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
rf.fit(train[predictors], train['target'])
```

```python
prediction = rf.predict(test[predictors])
```

```python
acc = accuracy_score(test['target'], prediction)
```

```python
# What was the performance (average) in the n previous matches 

def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')

    # Include only the past 3 days data
    rolling_stats = group[cols].rolling(3, closed= 'left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset= new_cols)
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = df.groupby('team', as_index= False).apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling.index = matches_rolling.index.droplevel(0)
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
# Retraining the model
def make_predictions(data, predictors):
    train = data[data['date'] < '2022-01-01']
    test = data[data['date'] >= '2022-01-01']

    rf.fit(train[predictors], train['target'])
    prediction = rf.predict(test[predictors])

    combined = pd.DataFrame(dict(actual= test['target'], prediction= prediction))
    combined = combined.merge(data[['date', 'team', 'opponent', 'result']], left_index= True, right_index= True)
    pd.crosstab(index=combined['actual'], columns= combined['prediction'])
    precision = precision_score(test['target'], prediction)

    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors+new_cols)
```
