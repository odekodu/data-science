# Computer Science

A computer is an electronic device that takes input from the user, processes it according to a set of instructions, outputs the data and also temporarily or permantently stores the data.

It is made up of Hardware and Software.

### Hardware
These are the physical components of the computer. eg. monitor, mouse, printer etc.

### Software
These are the components of the computer that perform specific tasks on a computer based on a collection of instructions. eg. whatsapp, facebook

There are two types of software; Application and System software.

`System software` acts as an intermediary between the user and the computer. It is designed to use and maintain the computer's hardware. eg. Operating Systems, Firmwares, Device Drivers.

`Appliation Software` is the software design to help users perform specific tasks on the computer. eg. Whatsapp Facebook etc.


# Computer Programming

`Computer Programming`, also known as coding, involves writing a well structured sets of instructions for a computer to execute.

`Computer Language`; Computers only understands binary (0, 1).

`Programming languages` can be divided into two
- Low-Level Languages
- High-Level Languages

`Low-level langeuges`
- Machine Language
- Assembly Language

`Machine Language`; This executes directly on the computer.

`Assembly Language`; This needs an assebler to be converted into a machine language before being executed by the computer.

`High-Level Languages`; The require compilers or interpreters to be converted into what the computer can execute.

`Compiler`; This converts the entire code at once to a language the computer can understand `(Faster than Interpreters)`

`Interpreter`; Converts one line at a time; each time the line is encountered. 


`Data`; This is a collection of values. It can be either:
- Digital Data
- Analog Data

`Digital Data`; They are programmed sequencially and jumps from one value to the next. They are discrete and discontinious.

`Analog Data`; They fluctuate infinitely with a specific range

`Analog Devices`; Speakers, Amplifiers

`Data` can be converted from digital to analog and vise-versa with use of `DAC` and `ADC` converters

`Information`; This is processed data.

`Bits and Bytes`: A bit (b) it the unit for measuring network speed and a byte (B) is unit of measure a data in a computer. `1B = 8b`.

`Variables`: Use the name values stored. They should be relatable.

`Datatype`: Used to mark the type of data so the computer know the operations to apply to the data.
eg. Number (Integer, Float), Text, Boolean, Null.

`Type Checking`; This checking for operations that doesn't belong to a data-type. It can be done during compilation or `run-time`
